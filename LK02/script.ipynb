```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

# 1 Kelas dan ngoding

```python
suka_ngoding = (df['Suka Ngoding'].value_counts()['Ya'])
tidak_suka = (df['Suka Ngoding'].value_counts()['Tidak'])
labels = ["Suka","Tidak Suka"]
data = [suka_ngoding,tidak_suka]
plt.pie(data,labels = labels,autopct='%1.1f%%',shadow=True)
plt.title("Jumlah responden PTI-C Kelas Statistika Dasar yang suka ngoding")
plt.show()


```

# Kelamin dan ngoding
