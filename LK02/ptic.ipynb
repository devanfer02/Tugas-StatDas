```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hasilsurvey.csv')
df.head()
```

# Sections

### Raion

#### Ingin Join Raion

```python
join_raion =  df['Ingin join Raion'].value_counts()['Ya']
tidak_join = df['Ingin join Raion'].value_counts()['Tidak']
labels_raion = ['Ingin Join','Tidak Ingin Join']
data_raion = [join_raion,tidak_join]
raion_colors = ['#0000FF','#FFFF00']
plt.pie(data_raion,labels=labels_raion,autopct='%1.1f%%',colors=raion_colors)
plt.title('Populasi PTI C Kelas Statistika Dasar yang ingin bergabung dengan Raion')
plt.show()
```

#### Tertarik Ngoding dan Ingin Join Raion

```python
raion_df = df[df['Ingin join Raion'] == 'Ya']
raion_suka_ngoding = raion_df['Suka Ngoding'].value_counts()['Ya']
raion_tidak_ngoding = raion_df['Suka Ngoding'].value_counts()['Tidak']
labels_raion_ngoding = ['Suka Ngoding','Tidak Suka']
data_raion_suka_ngoding = [raion_suka_ngoding,raion_tidak_ngoding]
plt.pie(data_raion_suka_ngoding,labels=labels_raion_ngoding,autopct='%1.1f%%',colors=raion_colors)
plt.title('Yang tertarik dengan Raion dan Ngoding')
plt.show()

```

### Kepribadian 

#### Banyaknya Introvert dan Ekstrovert

```python
kepribadian =  df.groupby(['Kepribadian'],as_index=False).count()
kepribadian
introvert = df['Kepribadian'].value_counts()['Saya Introvert']
ekstrovert = df['Kepribadian'].value_counts()['Saya Extrovert']
colors_kepribadian = ['#2c9892','#ffc94e']
data_kepribadian = [introvert,ekstrovert]
labels_kepribadian = ['Introvert','Ekstrovert']
plt.pie(data_kepribadian,labels=labels_kepribadian,autopct='%1.1f%%',colors=colors_kepribadian)
plt.title('Kepribadian Populasi Kelas PTI C')

plt.show()
```
